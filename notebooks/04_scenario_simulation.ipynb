# 04 — Scenario Simulation

"What-if" projections: savings increase, expense reduction, income change.
Uses simple projection logic — no ML.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))

import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from src.data_loader import load_data, preprocess, get_user_data
from src.scenario_simulation import (
    simulate_savings_increase, simulate_expense_reduction,
    simulate_income_change, compare_scenarios, project_balance
)
```

```python
df = preprocess(load_data('../data/virtual_financial_advisor_data.csv'))
user_df = get_user_data(df, 'user_1')
```

## 1. Scenario: Save 10% More

```python
s1 = simulate_savings_increase(user_df, increase_pct=10, months=12)
for k, v in s1.items():
    print(f'{k}: {v}')
```

## 2. Scenario: Reduce Dining by 20%

```python
s2 = simulate_expense_reduction(user_df, category='Dining', reduction_pct=20, months=12)
for k, v in s2.items():
    print(f'{k}: {v}')
```

## 3. Scenario: Income Increase of 15%

```python
s3 = simulate_income_change(user_df, change_pct=15, months=12)
for k, v in s3.items():
    print(f'{k}: {v}')
```

## 4. Compare Scenarios

```python
comparison = compare_scenarios([s1, s2, s3])
comparison
```

```python
# Projected vs baseline bar chart
fig = go.Figure()
for sc in [s1, s2, s3]:
    fig.add_trace(go.Bar(
        name=sc['scenario'],
        x=['Baseline', 'Projected'],
        y=[sc['baseline_cumulative'], sc['projected_cumulative']]
    ))
fig.update_layout(barmode='group', title='Cumulative Savings: Baseline vs Projected')
fig.show()
```

## 5. Balance Projection

```python
balance = project_balance(user_df, initial_balance=5000, months=12)
balance
```

```python
fig2 = px.line(balance, x='month', y='projected_balance',
               title='12-Month Balance Projection (Starting: $5,000)',
               labels={'projected_balance': 'Balance ($)', 'month': 'Month'})
fig2.show()
```
